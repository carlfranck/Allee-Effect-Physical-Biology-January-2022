```python
def CI_from_split_data(data,d):
    
    y_true = data[:,1] # y-data (log of density in cells/mL)
    x1_lr = data[:,0] # x1-data (time in hours) 
    x = data[:,0]
    nzeros = len([x[i] for i in range(len(x)) if x[i] < d])
    nones = len([x[i] for i in range(len(x)) if x[i] >= d])
    x2_lr = np.hstack((np.zeros(nzeros), np.ones(nones))) # x2-data (0 if before transition time 'd', 1 if after)

    X = np.column_stack((x1_lr, x2_lr)) # stacks data into columns
    out_df = pd.DataFrame(X, columns=['x1','x2']) # creates dataframe
    out_df['y'] = y_true # sets 'y'-data for dataframe
    Xb = sm.add_constant(out_df[['x1','x2']]) # adds categories to dataframe
    poly = PolynomialFeatures(interaction_only=True) # defines polynomial function
    X_tr = poly.fit_transform(Xb) # establishes how model handles dataframe data    
    # dataframe readjustment
    Xt = pd.concat([Xb,pd.DataFrame(X_tr,columns=poly.get_feature_names()).drop(['1','x1','x2','x0','x0 x1','x0 x2'],1)],1)   
    mod = sm.OLS(y_true, Xt) # defines model
    res = mod.fit() # fits model
    
    cis = res.conf_int(alpha=0.05, cols=None) # pulls out confidence interval data
    return [cis[0][3], cis[1][3]], res.params, [cis[0][1], cis[1][1]] # returns confidence interval of interest
```

```python
def boundary_select(data, d, c):
    
    analyze = CI_from_split_data(data,d)
    
    plt.scatter(data[:,0],data[:,1],s=2, color=c) # red
    plt.plot([d, d], [3.3,4.6], color=c, linestyle='--')
    m1 = analyze[1][1]
    b1 = analyze[1][0]
    plt.plot([0, d], [b1+0*m1,b1+d*m1],color=c, linestyle='-')
    m2 = analyze[1][1]+analyze[1][3]
    b2 = analyze[1][0]+analyze[1][2]
    plt.plot([d, 83], [b2+d*m2,b2+83*m2],color=c, linestyle='-')
    print(m1, 'Slope 1')
    print(analyze[2], 'CI for Slope 1')
    dm1 = (analyze[2][1]-analyze[2][0])/2
    print(dm1, 'Delta Slope 1')
    print(m2, 'Slope 2')
    print(analyze[2], 'CI for (Slope 2)-(Slope 1)')
    dm1mm2 = (analyze[0][1]-analyze[0][0])/2
    print(dm1mm2, 'Delta (Slope 2)-(Slope 1)')
    
    dm2 = np.sqrt(dm1**2+dm1mm2**2)
    print(dm2, 'Delta Slope 2')
    
    print(m2/m1, 'Ratio of (Slope 2)/(Slope 1)')
    print((m1/m2)*np.sqrt((dm1/m1)**2+(dm2/m2)**2), 'Delta Ratio of (Slope 2)/(Slope 1)')
    

    plt.ylim([3.3,4.6])
    plt.xlim([0,83])
    plt.grid()
    #plt.show()
    
    return analyze[0]
```
